```python
import numpy as np
import networkx as nx
from pysmiles import read_smiles
```

```python
def sm2graph(smiles,size, weight = None):
    mol = read_smiles(smiles)
    #adjacent matrix
    adj = nx.adjacency_matrix(mol, weight=weight).todense().A
    adj = adj+np.eye(adj.shape[0])
    adj = np.pad(adj,(0,size-adj.shape[0]))
    #degree matrix
    de = np.zeros((size,size))
    for i in mol.degree:
        de[i[0]][i[0]] = i[1]
    mole = mol.nodes(data='element')
    return adj, de, mole
```

```python
def process_data(filefolder):
    periodic_table = ('H', 'Li', 'Be', 'B', 'C', 'N', 'O', 'F', 'Na', 'Mg', 'Al', 'Si', 'P', 'S', 'Cl', 'K', 'Ca', 'Sc', 'Ti', 'V', 'Cr', 'Mn', 'Fe', 'Co', 'Ni', 'Cu', 'Zn', 'Ga', 'Ge', 'As', 'Se', 'Br', 
                  'Rb', 'Sr', 'Y', 'Zr', 'Nb', 'Mo', 'Te', 'Ru', 'Rh', 'Pd', 'Ag', 'Cd', 'In', 'Sn', 'Sb', 'Te', 
                  'I', 'Cs', 'Ba', 'La', 'Ce', 'Pr', 'Nd', 'Pm', 'Sm', 'Eu', 'Gd', 'Tb', 'Dy', 'Ho', 'Er', 'Tm', 
                  'Yb', 'Lu', 'Hf', 'Ta', 'W', 'Re', 'Os', 'Ir', 'Pt', 'Au', 'Hg', 'Tl', 'Pb', 'Bi', 'Po', 'At', 
                  'Fr', 'Ra', 'Ac', 'Th', 'Pa', 'U', 'Np', 'Pu', 'Am', 'Cm', 'Bk', 'Cf', 'Es', 'Fm','Md', 'No', 'Lr',
                  'Rf', 'Db', 'Sg', 'Bh', 'Hs', 'Mt', 'Ds', 'Rg', 'Cn', 'Nh', 'Fl', 'Mc', 'Lv', 'Ts', 'Og', 'Uue')
    filefolder = '../'+filefolder
    data_path = filefolder + '/names_smiles.csv'
    data_adj = []
    data_de = []
    feature = []
    with open(data_path,'r') as f:
        header = f.readline().replace('\n','').split(',')
        if header[0] == 'SMILES':
            index = 0
        else:
            index = 1
        for line in f.readlines():
            line = line.replace('\n','').split(',')
            smiles = str(line[index])
            adj,de, mole = sm2graph(smiles,78)
            data_adj.append(adj)
            data_de.append(de)
            onehot = np.zeros((78,113))
            i=0
            for atom in mole:
                for j in range(113):
                    print(periodic_table[j])
                    if atom[1] == periodic_table[j]:
                        onehot[i][j] = 1
                        break
                i = i+1
            feature.append(onehot)
                        
    np.save(filefolder+'/adj_matrix.npy',np.array(data_adj))
    np.save(filefolder+'/de_matrix.npy',np.array(data_de))
    np.save(filefolder+'/feature.npy',np.array(feature))   
```

```python
def load_data(filefolder):
    ori_filefolder = filefolder
    filefolder = '../'+filefolder
    adj_filename = filefolder + '/adj_matrix.npy'
    de_filename = filefolder + '/de_matrix.npy'
    feature_filename = filefolder + '/feature.npy'
    adj = np.load(adj_filename)
    de = np.load(de_filename)
    feature = np.load(feature_filename)
    if ori_filefolder == 'test':
        label_filename = filefolder + '/output_sample.csv'
    else:
        label_filename = filefolder + '/names_labels.csv'
#         label = pd.read_csv(os.path.abspath(filefolder + '/names_labels.csv'), sep=',')
    label = []
    with open(label_filename,'r') as f:
        header = f.readline().replace('\n','').split(',')
        if header[0]=='Label':
            label_index = 0
        else:label_index = 1
        for line in f.readlines():
            line = line.replace('\n','').split(',')
            label.append(int(line[label_index]))
    label = np.array(label)
    return adj, de, feature ,label
```
